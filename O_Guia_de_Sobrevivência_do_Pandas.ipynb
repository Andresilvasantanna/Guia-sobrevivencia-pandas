### 1.1.2 - O que é pandas

O pandas é uma biblioteca do Python voltada para manipulação e análise de dados estruturados.

Sua principal estrutura é o DataFrame, uma tabela bidimensional composta por colunas chamadas Series. Cada Series funciona como um vetor de dados associado a um índice, permitindo operações vetorizadas rápidas e eficientes.

Diferente de planilhas tradicionais, o Pandas não é orientado a células, mas sim a transformações aplicadas sobre conjuntos de dados (vetores/colunas). Isso permite realizar filtragens, agregações, tratamentos de dados ausentes e análises estatísticas de forma escalável e expressiva.

Internamente, o Pandas utiliza amplamente o NumPy para operações vetorizadas e organização eficiente dos dados em memória.

***Documentação:*** (https://pandas.pydata.org/)

#### 1.1.2.1 Vetor
O vetor é uma sequência organizada de valores sobre a qual é possível aplicar operações de forma simultânea.

**Exemplo:**

`[22, 38, 26, 35]`

Em Pandas, uma coluna como:

`df['Age']`

pode ser entendida como um vetor de idades.

A principal característica de um vetor é permitir operações vetorizadas, ou seja, operações aplicadas automaticamente a todos os elementos do conjunto sem necessidade de loops explícitos.

**Exemplo:**

`df['Age'] + 10`

**Resultado conceitual:**

`[22, 38, 26] → [32, 48, 36]`

Assim, no Pandas, os dados deixam de ser tratados célula por célula e passam a ser manipulados como estruturas vetoriais organizadas em colunas (Series).

#### 1.1.2.1 - Series

No pandas, uma Series é uma estrutura unidimensional composta por:

- um vetor de dados
- um índice associado a cada valor

Ela pode ser entendida como a representação de uma única coluna de um DataFrame.

**Exemplo conceitual:**

|Indice|Valor|
|------|-----|
|0|22|
|1|38|
|2|26|


Nesse caso:

- à esquerda está o índice
- à direita estão os valores do vetor

Quando executamos:

`df['Age']`

o retorno é uma Series.

Uma Series permite:

- operações vetorizadas
- alinhamento automático por índice
- filtragens e transformações eficientes

**Estruturalmente:**

Series = índice + vetor de dados

E um DataFrame pode ser entendido como um conjunto de Series alinhadas pelo mesmo índice.

#### 1.1.2.3 - Colunas

No pandas, uma coluna representa uma Series dentro de um DataFrame.

Cada coluna possui:

- um nome (label)
- um índice compartilhado com o restante do DataFrame
- um vetor de dados associado

Exemplo:

`df['Age']`

Nesse caso:

- Age é o nome da coluna
- o retorno é uma Series

Estruturalmente:


```text
DataFrame
 ├── Coluna Age  → Series
 ├── Coluna Fare → Series
 └── Coluna Sex  → Series
```

As colunas são a principal unidade de trabalho no Pandas, permitindo operações vetorizadas, filtragens, agregações e transformações de dados de forma eficiente.

#### 1.1.2.4 - DataFrame

O DataFrame é uma estrutura tabular bidimensional composta por múltiplas Series alinhadas por um mesmo índice.
Cada coluna do DataFrame é uma Series, e cada Series contém:

- um vetor de dados


- um índice associado

Exemplo estrutural:

```
DataFrame
 ├── Coluna Age  → Series
 ├── Coluna Fare → Series
 └── Coluna Sex  → Series
 ```

O DataFrame é a principal estrutura do Pandas para manipulação e análise de dados, permitindo:


- operações vetorizadas


- filtragens

- agregações

- transformações
- alinhamento automático por índice

Diferente de planilhas tradicionais, o Pandas trabalha orientado a colunas (Series) e operações sobre vetores, e não célula por célula.

### 1.4.5 df.query
Embora será possível fazer filtros usando condicionais, o código rapidamente se torna complexo . Nesse cenário que temos a alternativa de usar o método `query()`

**O que é df.query()?**

O método `df.query()` permite filtrar um DataFrame usando expressões em formato de texto, quase como uma mini linguagem de consulta.

**Comparação**
|Forma tradicional| Df Query|
|-----------------|---------|
|df.loc[df['Age'] > 18]|df.query("Age > 18")|
|df.loc[(df['Age'] > 18) <br> &(df['Fare'] > 50)]| df.query("Age > 18 and Fare > 50")|

##### **Notas**
1. Dentro do query usamos ***and*** e **or**
2. Na forma clássica usamos ***&*** e **|**

**Documentação:** (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.query.html)


````python
#comparação
df.query("Age > 18")
````

````python
#igualdade
df.query("Sex == 'female'")
````

````python
#multiplas condições
df.query("Sex == 'female' and Pclass == 1")
````

````python
#selecionando colunas após o df.query()
df.query("Age > 18")[['Name', 'Age', 'Fare']]
````

#### 1.4.5.1 - Variáveis

Uma das maiores vantagens do método `query()` é a possibilidade de utilizar variáveis externas do Python dentro da expressão de filtragem.

Com isso, o código ganha:

- legibilidade
- clareza
- facilidade de manutenção
- reutilização de filtros dinâmicos

As variáveis externas são referenciadas utilizando o operador `@`.

Exemplo:

```python
idade = 30

df.query("Age > @idade")
```

````python
idade = 25
df.query("Age > @idade")
````

````python
sexo = 'male'
df.query(" Age>= @idade and Sex == @sexo")
````

````python
#Filtros multiplos
# após retorna as colunas necessárias

idade_min = 18
idade_max = 40
sexo = 'female'
classes = [1, 2]
tarifa = 50

df.query(
    "(@idade_min <= Age <= @idade_max) "
    "and Sex == @sexo "
    "and Pclass in @classes "
    "and Fare > @tarifa"
)[['Name', 'Age', 'Pclass', 'Fare']]

#note que, embora tenha mais linhas de código, a sintaxe é mais fluída que a forma tradicional:
#df.loc[
#    (df['Sex'] == sexo) &
#    (df['Pclass'].isin(classes)) &
#    (df['Fare'].between(50, 150))
#]
````

## 1.5 - Funções

### 1.5.1 - Apply()

**O que é apply()?**

O método: `apply()`permite aplicar uma função sobre:
- uma Series
- ou sobre linhas/colunas de um DataFrame




````python
df['Age'].apply(lambda x: x + 10)
````

**O que acontece?**

A coluna Age é uma Series.

O apply():
- pega cada valor
- chama de x

executa: `x + 10`

**Importante**

Isso: `df['Age'] + 10` faz a mesma coisa e normalmente é:
- mais rápido
- mais idiomático
- mais “Pandas

**Primeira grande lição do apply**
  ❌ Não use apply() para operações vetorizáveis simples.

**Então quando ele começa a brilhar?**

Quando sua lógica:
- é complexa
- não existe pronta no Pandas
- exige função customizada

Abaixo um exemplo de uso que exige apply

````python
def categoria_idade(x):
    if x < 18:
        return 'Menor'
    elif x < 60:
        return 'Adulto'
    else:
        return 'Idoso'
````

````python
df['FaixaEtaria'] = df['Age'].apply(categoria_idade)
````

**Considerações:**

O que o apply faz aqui?
Para cada idade:
- chama a função
- recebe o retorno
- monta uma nova Series

Ou seja, o apply():
- recebe uma função
- executa várias vezes
- constrói o resultado fina

Isso significa que em Series o `apply()` opera elemento foi elemento

Mas em DataFrame:
- pode operar por coluna
- ou por linha

Para isso, precisamos passar o argumento **axis**

**Axis**

O axis não significa "horizontal" ou "vertical" e sim  a direção da operação

| Axis     | Significado                 |
| -------- | --------------------------- |
| `axis=0` | operar AO LONGO das linhas  |
| `axis=1` | operar AO LONGO das colunas |

Exemplo

|Indice|Age| Fare|
|------|---|-----|
|0|22|7.25|
|1|38|71.83|
|2|26|7.92|

**Axis = 0**

O pandas percorre as colunas ↓
```
Exemplo:
df[['Age', 'Fare']].apply(max, axis=0)
```

```
Age     80
Fare   512
```
**axis=1**

O Pandas percorre as linhas →
```
Exemplo
df[['Age', 'Fare']].apply(sum, axis=1)
```

Para cada linha: Age + Fare


⚠️ Aqui está o ponto CRÍTICO

No axis=0, a função recebe Series de coluna

No axis=1, a função recebe uma Series de linhas

Exemplos:
```
def exemplo(x):
    print(type(x))
```

Com axis=0 , cada x será um coluna interira.

Ex: `df.apply(exemplo, axis=0)`

Com axis=1, cada x será uma linha inteira.

Ex: `df.apply(exemplo, axis=1)`

**Resumo**

| Axis     | A função recebe       |
| -------- | --------------------- |
| `axis=0` | uma coluna (`Series`) |
| `axis=1` | uma linha (`Series`)  |






### 1.5.2 - Diferença entre aplicar em colunas e linhas
#### 1.5.2.1 - Aplicar em colunas (axis=0)

O Pandas entrega uma coluna inteira (Series) por vez.

`df[['Age', 'Fare']].apply(max, axis=0)`

O que a função recebe?
- Perimeiro , a  _Series Age_ e depois a _Series Fare_

Então o foco aqui é operações sobre colunas inteiras

Exemplos:
- média
- máximo
- mínimo
- padronização
- limpeza de texto

### 6.2.2 -  Aplicar em linhas (axis=1)

Agora o Pandas entrega uma linha inteira (Series) por vez.

Exemplo: `df[['Age', 'Fare']].apply(sum, axis=1)`

**Exemplo de uso:**
```
def categoria(row):
    if row['Age'] < 18:
        return 'Menor'
    elif row['Fare'] > 100:
        return 'VIP'
    else:
        return 'Adulto'
```

Aplicação: `df.apply(categoria, axis=1)`


Para cada linha:
- a função recebe a linha inteira
- acessa:
```
row['Age']
row['Fare']
```

Em Axis = 0 , pensamos em **coluna → transformação**

Em Axis = 1 , pensamos em **registro/linha → decisão**


**Conclusão**

Use axis=0, quando:
- trabalha com uma variável isolada
- estatísticas
- limpeza
- transformação vetorizada

Use axis=1 quando:
- precisa combinar colunas
- criar regras condicionais
- tomar decisões por registro


⚠️ ALERTA MUITO IMPORTANTE

O axis=1 costuma ser:
- mais lento
- menos vetorizado
- mais próximo de loop


**Nota:**

Sempre se pergunte: “consigo resolver isso sem axis=1?”

Porque muitas vezes:
- np.where
- máscaras booleanas
- vetorização

são melhores.

### 1.6.1 -  assign()

O método `assign()`permite criar colunas sem modificar diretamente o dataframe original

**Comparação**

Forma tradicional: `df['Age_Double'] = df['Age'] * 2`
com assign:
```
df.assign(
    Age_Double = df['Age'] * 2
)
```

O método tradicional modifica o DataFrame, já o `assign()`retorna um novo dataframe

isso é importante no pipelines de dados pois podemos criar novas colunas sem modificar o dataframe original

**Assign()**
Criando múltiplas colunas
```
df.assign(
    Age_Double = df['Age'] * 2,
    Adulto = df['Age'] >= 18
)
```

Isso é diferente do apply()

No apply(), o x pode ser:
- elemento
- linha
- coluna

No assign(), o lambda recebe o DataFrame inteiro

**Vantagens**
Você consegue:
- criar pipelines limpos
- evitar variáveis intermediárias
- escrever código mais fluido

**Quando usar assign()**
- pipelines
- notebooks organizados
- method chaining
- múltiplas transformações

**Atribuição tradicional**

Melhor para:
- exploração rápida
- scripts simples
- debugging

**Extra: método chaining**

Chaining vem da ideia de encadear operações ou criar uma cadeia de métodos.

```
(

O que acontece
```

### 1.6.2 - múltiplas condições com np.select
o O np.where() funciona muito bem para `SE / SENÃO`

Mas e quando você precisa de:
```SE
SENÃO SE
SENÃO SE
SENÃO
```
Aí entra o `np.select()` pois ele permite múltiplas condições vetorizadas.
A sintaxe é
```
np.select(
    condicoes,
    escolhas,
    default
)
```
**Exemplo 1**

```
import numpy as np

condicoes = [
    df['Age'] < 18,
    df['Age'] < 60
]

escolhas = [
    'Menor',
    'Adulto'
]

df['FaixaEtaria'] = np.select(
    condicoes,
    escolhas,
    default='Idoso'
)
```

**Exemplo 2**
```
condicoes = [
    df['Fare'] > 100,
    df['Fare'] > 50,
    df['Fare'] > 20
]

categorias = [
    'Luxo',
    'Premium',
    'Intermediário'
]

df['CategoriaTarifa'] = np.select(
    condicoes,
    categorias,
    default='Econômica'
)
```
Isso é MUITO usado em:
- feature engineering
- segmentação
- scoring
- classificação
- regras de negócio

**Comparação**

| Ferramenta    | Melhor uso                  |
| ------------- | --------------------------- |
| `np.where()`  | duas opções                 |
| `np.select()` | múltiplas regras            |
| `apply()`     | lógica customizada complexa |

**Nota**

Muita gente usa `apply(axis=1)` para lógica condicional. No entanto, poderia usar `np.select()`

O `np.select()` é :
- mais rápida
- mais vetorizada
- mais elegante



````python
import numpy as np

condicoes = [
    df['Age'] < 18,
    df['Age'] < 60
]

escolhas = [
    'Menor',
    'Adulto'
]

df['FaixaEtaria'] = np.select(
    condicoes,
    escolhas,
    default='Idoso'
)
````

````python
condicoes = [
    df['Fare'] > 100,
    df['Fare'] > 50,
    df['Fare'] > 20
]

categorias = [
    'Luxo',
    'Premium',
    'Intermediário'
]

df['CategoriaTarifa'] = np.select(
    condicoes,
    categorias,
    default='Econômica'
)
````

### 1.6.3 - map()

O `.map()` é usado principalmente para transformar valores de uma Series através de um mapeamento.

**A ideia central é** valor antigo → valor novo

**Exemplo clássico**
```
df['Sex'].map({
    'male': 'Homem',
    'female': 'Mulher'
})

```
O que acontece?
O Pandas percorre a Series:
```
male
female
male
```

e substitui por:
```
Homem
Mulher
Homem
```
Isso é um mapeamento chave → valor. Esse tipo de mapeamente é muito usado para:

### 1.7.1 - Fundamentos do Groupby()

**Princípios**

**Groupby**: a ideia desse método é dividir os dados em grupos com base nos critérios selecionados e em seguida, aplicar uma operação a esses grupos
Esse método possui diversos parâmetros, sendo os principais:


**by**: especifica a coluna ou colunas que desejamos agrupar. Como argumento podemos passar o nome de uma coluna oou uma lista com os nomes das colunas


**axis**: especifica o eixo ao longo do qual(is) desejamos agrupar. O valor padrão é 0. O que siginifica que as linhas serão agrupadas. Se quisermos agrupar a colunas, devemos especificar axis = 1.


**sort**: parâmetro booleano( True ou False) que indica se os grupos devem ser ordenados de maneira ascendente ou descendentes. o valor padrão é True.


**dropna**:  utilizamos o dropana para controlar se os valores nulos ( NaN) serão mantidos ou excluídos durante o processo de agrupamento. o padrão é True.

**Mentalidade correta:** dividir → aplicar → combinar



````python
df.groupby('Pclass')['Age'].mean()
 #“agruppe os passageiros por classe e calcule a média da idade em cada grupo”
````

Para cada grupo o `mean()` na coluna **Age**

### 7.1.4 - transform()

O método transform() permite executar uma função para cada valor do DataFrame.

**sintaxe**

dataframe.transform(func, axis, raw, result_type, args, kwds)

**No entanto**, ele devolve resultado no tamanho original do DataFrame.

**Comparação**
- df.groupby('Pclass')['Fare'].mean()

    | Pclass | Fare |
    | ------ | ---- |
    | 1      | 84   |
    | 2      | 20   |
    | 3      | 13   |

  890 linhas → 3 grupos

- df.groupby('Pclass')['Fare'].transform('mean')
  Saída: Uma Series com 890 linhas
  Exemplo
  | Passenger | Pclass | Fare | Transform mean |
  | --------- | ------ | ---- | -------------- |
  | A         | 1      | 80   | 84             |
  | B         | 1      | 100  | 84             |
  | C         | 2      | 20   | 20             |

**O que aconteceu?**
Cada passageiro recebeu a média da sua própria classe

🔥Vantagens

- comparar indivíduo vs grupo
- criar métricas contextuais
- fazer feature engineering




````python
df['Fare_Diff'] = (
    df['Fare']
    - df.groupby('Pclass')['Fare'].transform('mean')
)
````

````python
df.head(10)
````

Isso é MUITO usado em:
- machine learning
- detecção de outliers
- análise contextual
- normalização

⚠️ Insight MAIS importante

O transform() NÃO reduz.
| Método        | Retorno                 |
| ------------- | ----------------------- |
| `agg()`       | reduz grupos            |
| `transform()` | mantém tamanho original |
| `apply()`     | flexível                |

###3.2 Concat()

O concat() é um método do pandas usado para unir DataFrames ou Series de forma estrutural, empilhando os dados verticalmente (axis=0) ou horizontalmente (axis=1).

Diferente do merge(), que relaciona tabelas por chaves em comum, o concat() simplesmente combina estruturas mantendo seus dados originais e alinhando-os pelos índices quando necessário.

Ele é muito utilizado para juntar arquivos mensais, consolidar múltiplos datasets, empilhar tabelas semelhantes ou adicionar colunas lado a lado em processos de ETL e análise de dados.

**Sintaxe básica:** `pd.concat([df1, df2])`

**Documentação:**: https://pandas.pydata.org/docs/reference/api/pandas.concat.html

#### 3.2.1 Concat vertical (axis=0)

Nesse caso, o pandas empilha uma tabela abaixo da outra

Como as colunas são diferentes o Pandas cria a união das colunas

E onde faltar dado?
- aparece NaN

O concat() não faz relacionamento

````python
pd.concat([
    passageiros,
    embarque
])
````

#### 3.2.2 Concat horizontal (axis=1)

````python
pd.concat([
    passageiros,
    embarque
], axis=1)
````

**Diferença essencial**
- merge(): “relacione os dados”
- concat(): “cole as estruturas”

Índices no concat

O alinhamento ocorre pelos índices
**Exemplo**

Se:passageiros.index

↓

`0 1 2 3`

e:

embarque.index

↓

`0 1 2 3`
O concat horizontal funciona perfeitamente.

Mas se os índices forem diferentes:

o Pandas tentará alinhar por posição do índice e não pela ordem visual.

**Muito importante:** `ignore_index=True`

````python
pd.concat([
    passageiros,
    embarque
], ignore_index=True)
````

**O que faz?**

Recria um novo índice sequencial

Isso é muito usado quando:
- empilhamos arquivos
- juntamos meses
- consolidamos datasets

````python
pd.concat(
    [passageiros, embarque],
    keys=['passageiros', 'embarque']
)
````

Outro parâmetro útil :keys=

Nesse caso é criado um multindex que é muito usado para:
- rastrear origem
- auditoria

````python
pd.concat(
    [passageiros, embarque],
    keys=['passageiros', 'embarque']
)
````

## 3.3 combine_first()

O combine_first() combina dois DataFrames priorizando os valores do primeiro e utilizando os valores do segundo apenas onde existirem dados ausentes (NaN). Ele é muito útil em processos de limpeza, consolidação e recuperação de dados incompletos, funcionando como uma espécie de “fallback” entre estruturas compatíveis.

````python
passageiros_nan = passageiros.copy()
````

````python
# remomvendo alguns valores
passageiros_nan.loc[
    [0, 3, 5],
    'Age'
] = np.nan
````

````python
resultado = passageiros_nan.combine_first(
    idades_backup
)
resultado
````

**Diferença para fillna**

O fillna() normalmente usa:
- valor fixo
- média
- constante

`df['Age'].fillna(0)`

Já o combine_first() usa dados reais de outro DataFrame

Isso é MUITO útil em:
- ETL
- reconciliação
- múltiplas fontes
- bases complementares

## 3.4 Update()

O update() modifica um DataFrame existente utilizando valores de outro DataFrame alinhados por índice e coluna. Diferente do merge() ou concat(), ele não adiciona novas linhas ou colunas; apenas atualiza valores já existentes, sendo muito útil para correções, patches e sincronização de dados.

**Ideia central:** atualizar valores existentes usando outro DataFrame

**Diferença IMPORTANTE**

- O combine_first() preenche apenas NaN
- O update() sobrescreve valores existentes

 **⚠️Muito importante**

O update() modifica o DataFrame original Diferente de muitos métodos do Pandas.

O update() faz mutação inplace

**Então:** `resultado = passageiros.update(...)` retornará None.
Porque ele altera diretamente passageiros

**Outro detalhe importante**

O update() não cria novas linhas nem novas colunas
